# Get advanced stats per season per player

```python
def get_advanced_by_season(id_list):
    
    advanced = playerdashboardbyyearoveryear.PlayerDashboardByYearOverYear(player_id=str(id_list[0]), measure_type_detailed= 'Advanced', season_type_playoffs='Regular Season', per_mode_detailed='Per Game')
    advanced_df = advanced.get_data_frames()[1]
    name = [players.find_player_by_id(id_list[0])['full_name']]*len(advanced_df)
    advanced_df.insert(0, 'full_name', name)
    
    for player_id in id_list:
        next_player = playerdashboardbyyearoveryear.PlayerDashboardByYearOverYear(player_id=str(player_id), measure_type_detailed= 'Advanced', season_type_playoffs='Regular Season', per_mode_detailed='Per Game')
        next_player_df = next_player.get_data_frames()[1]
        name = [players.find_player_by_id(player_id)['full_name']]*len(next_player_df)
        next_player_df.insert(0, 'full_name', name)
        
        pd.concat([advanced_df,next_player_df], axis=1)
        
    return advanced_df
```
